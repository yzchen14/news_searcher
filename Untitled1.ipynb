```python
def convert_relative_date(raw_date):
    """Converts relative time strings like '13 å°æ™‚å‰' to 'YYYY-MM-DD'."""
    now = datetime.datetime.now()

    if "hours ago" in raw_date:
        hours_ago = int(re.search(r'(\d+)', raw_date).group(1))
        date_time = now - datetime.timedelta(hours=hours_ago)
    elif "minutes ago" in raw_date:
        minutes_ago = int(re.search(r'(\d+)', raw_date).group(1))
        date_time = now - datetime.timedelta(minutes=minutes_ago)
    elif "days ago" in raw_date:
        days_ago = int(re.search(r'(\d+)', raw_date).group(1))
        date_time = now - datetime.timedelta(days=days_ago)
    else:
        return raw_date  # If it's already absolute, return as-is

    return date_time.strftime("%Y-%m-%d")  # Convert to YYYY-MM-DD
```

```python
class NewsStorage:
    """Handles storing news articles in separate folders with details saved in text files."""
    
    def __init__(self, base_dir="NewsArchive"):
        self.base_dir = base_dir
        os.makedirs(self.base_dir, exist_ok=True)

    def save_news(self, news_item):
        """Creates a uniquely named JSON file for each news article and saves details."""
        filename = self.get_unique_filename(news_item["title"], news_item["date"])
        file_path = os.path.join(self.base_dir, filename + ".json")
        
        with open(file_path, "w", encoding="utf-8") as file:
            json.dump(news_item, file, ensure_ascii=False, indent=4)

    def get_unique_filename(self, title, date):
        """Generates a unique filename using title, date, and a hash."""
        sanitized_title = self.sanitize_filename(title)
        unique_id = hashlib.md5(title.encode()).hexdigest()[:6]  # Short hash for uniqueness
        return f"{date}-{unique_id}"

    @staticmethod
    def sanitize_filename(title):
        """Removes invalid characters from folder names to avoid file system errors."""
        return re.sub(r'[<>:"/\\|?*]', '', title)[:50]  # Limit name length to 50 characters


def convert_relative_date(raw_date):
    """Converts relative time strings like '13 å°æ™‚å‰' to 'YYYY-MM-DD'."""
    now = datetime.datetime.now()

    if "å°æ™‚å‰" in raw_date:
        hours_ago = int(re.search(r'(\d+)', raw_date).group(1))
        date_time = now - datetime.timedelta(hours=hours_ago)
    elif "åˆ†é˜å‰" in raw_date:
        minutes_ago = int(re.search(r'(\d+)', raw_date).group(1))
        date_time = now - datetime.timedelta(minutes=minutes_ago)
    elif "å¤©å‰" in raw_date:
        days_ago = int(re.search(r'(\d+)', raw_date).group(1))
        date_time = now - datetime.timedelta(days=days_ago)
    else:
        return raw_date  # If it's already absolute, return as-is

    return date_time.strftime("%Y-%m-%d")  # Convert to YYYY-MM-DD
```

```python
storage = NewsStorage()
```

```python
keywords = ['EUV China', 'EUV samsung', 'EUV Intel']
by_pass_site = ['ettoday', 'storm.mg']
```

```python
news_results = []
for keyword in keywords:
    driver.get(f"https://www.google.com/search?q={keyword}&gl=us&tbm=nws&lr=lang_en&tbs=qdr:w")
    time.sleep(5)
    while True:
        soup = BeautifulSoup(driver.page_source, "html.parser")
        for el in soup.select("div.SoaBEf"):
            raw_date = el.select_one(".LfVVr")
            formatted_date = convert_relative_date(raw_date.get_text().strip()) if raw_date else "Unknown"
            link = el.find("a")["href"]

            by_pass = False
            for bypass in by_pass_site:
                if bypass in link:
                    by_pass = True
                    break
            if by_pass:
                continue
    
            news_item = {
                    "link": link,
                    "title": el.select_one("div.MBeuO").get_text(),
                    "snippet": el.select_one(".GI74Re").get_text(),
                    "date": formatted_date,
                    "source": el.select_one(".NUnG9d span").get_text(),
                    "date": formatted_date
        
                }
            news_results.append(news_item)
            storage.save_news(news_item)
    
        # Try to find and click the "Next" button
        try:
            next_button = driver.find_element(By.XPATH, '//*[@id="pnnext"]')  # Adjust XPath if needed
            next_button.click()
            print("Go to next page")
            time.sleep(3)  # Wait for the next page to load
        except Exception as e:
            
            print("No more pages found.")
            break  # Exit loop if no "Next" button exists
```
